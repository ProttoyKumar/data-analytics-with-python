```python
sales_q1 = np.array([
    [120, 150, 180],  # Product A
    [200, 170, 190],  # Product B
    [140, 160, 130]   # Product C
])

sales_q2 = np.array([
    [135, 165, 195],  # Product A
    [210, 180, 205],  # Product B
    [150, 175, 145]   # Product C
])
```

```python
sales_q1.ndim
```

```python
sales_q1.shape
```

```python
# Element-wise operations
# addition
sales_q1 + sales_q2
```

```python
sales_q2 -sales_q1 
```

```python
np.round((sales_q2-sales_q1)/sales_q1*100, 2)
```

```python
price_per_unit = np.array([
    [10.5, 11.0, 10.8],  # Product A
    [15.2, 14.8, 15.5],  # Product B
    [8.5,  8.8,  9.0]    # Product C
])
```

```python
rev_q1 = sales_q1 * price_per_unit
rev_q2 = sales_q2 * price_per_unit
```

```python
q1_discount = rev_q1 * 0.2 # 20% flat discount
q2_discount = rev_q2 * 0.2
```

```python
net_rev_q1 = rev_q1 - q1_discount
net_rev_q2 = rev_q2 - q2_discount
```

```python
net_rev_q1
```

```python
# total discounts in q1
np.sum(q1_discount)
```

```python
np.sum(net_rev_q1) # net revenue in q1
```

This returns the total net revenue across all products and regions in Q1.

```python
# customer id, name
a = np.array([['101', 'Mira'],
              ['102', 'Abdul'],
              ['103', 'Andrea']
              ])

# ID, purchase amount, purchase date
b = np.array([['101', 250.5, '2023-08-01'],
              ['102', 150.0, '2023-08-02'],
              ['103', 300.75, '2023-08-01']
              ])
```

```python
b = b[:, 1:3]
```

```python
np.hstack((a, b))
```

This simply stacks arrays horizontally (column-wise).

```python
# customer id, name
c = np.array([['104', 'Venkat'],
              ['105', 'John'],
              ['106', 'Kathy']
              ])
```

```python
np.vstack((a,c))
```

This simply stacks arrays vertically (row-wise).
