```python
x = np.array([[10,12,9],[15,11,13]])
np.sort(x, axis=None) # sorts all elements of the array as if it were flattened into a 1D array
```

np.sort(array, axis=-1) sorts the array along the last axis.
For a 2D array, axis=-1 is equivalent to axis=1, meaning each row is sorted independently.

```python
np.sort(x, axis=-1)
```
